```python
from sklearn.preprocessing  import LabelEncoder

feature = ['Gender','Region','Smoking_History','Diabetes_History','Hypertension_History','Physical_Activity','Diet_Quality','Alcohol_Consumption','Family_History',
           'Heart_Attack_Occurrence']

for i in feature:
    data[i] = LabelEncoder().fit_transform(data[i])

data.head()
```

Model selection

Logestic regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler


independent = data.drop('Heart_Attack_Occurrence',axis=1)
dependent = data['Heart_Attack_Occurrence']

x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.3, random_state=42)

standard = StandardScaler()
x_train_stan = standard.fit_transform(x_train)
x_test_stan = standard.transform(x_test)


model = LogisticRegression(class_weight='balanced')
model.fit(x_train_stan,y_train)

pred = model.predict(x_test_stan)

print(pred) 

accuracy = accuracy_score(y_test,pred)

classificationreport = classification_report(y_test,pred) 

print(f'accuarcy score of model : {accuracy:.2f}')

print('classification report of model :\n',classificationreport)

```
